```python

import math, random, time
import matplotlib.pyplot as plt

# ---------- Problem / Board / Component Specs ----------
BOARD_W, BOARD_H = 50, 50  # grid units
BOARD_CENTER = (BOARD_W/2.0, BOARD_H/2.0)

# Components and their base sizes (w,h). Rotation allowed 0 or 90 degrees.
COMP_SIZES = {
    'USB_CONNECTOR': (5, 5),
    'MICROCONTROLLER': (5, 5),
    'MIKROBUS_CONNECTOR_1': (5, 15),
    'MIKROBUS_CONNECTOR_2': (5, 15),
    'CRYSTAL': (5, 5),
}


CRYSTAL_MAX_DIST = 10.0
COM_MAX_DIST = 2.0
KEEPOUT_WIDTH = 10
KEEPOUT_DEPTH = 15

GRID_MIN_X, GRID_MIN_Y = 0, 0
GRID_MAX_X, GRID_MAX_Y = BOARD_W, BOARD_H

def rect_from_xywh(x, y, w, h):
    return {'x': x, 'y': y, 'w': w, 'h': h}

def center_of(rect):
    return (rect['x'] + rect['w']/2.0, rect['y'] + rect['h']/2.0)

def rects_overlap(a, b):
    return not (a['x'] + a['w'] <= b['x'] or b['x'] + b['w'] <= a['x']
                or a['y'] + a['h'] <= b['y'] or b['y'] + b['h'] <= a['y'])

def inside_board(rect):
    return rect['x'] >= 0 and rect['y'] >= 0 and rect['x'] + rect['w'] <= BOARD_W and rect['y'] + rect['h'] <= BOARD_H

def distance(p, q):
    return math.hypot(p[0]-q[0], p[1]-q[1])


def seg_intersects_rect(A, B, rect):
    # If either endpoint inside rect -> intersection
    if (rect['x'] <= A[0] <= rect['x']+rect['w'] and rect['y'] <= A[1] <= rect['y']+rect['h']) \
       or (rect['x'] <= B[0] <= rect['x']+rect['w'] and rect['y'] <= B[1] <= rect['y']+rect['h']):
        return True

    rect_edges = [
        ((rect['x'], rect['y']), (rect['x']+rect['w'], rect['y'])),  # top
        ((rect['x']+rect['w'], rect['y']), (rect['x']+rect['w'], rect['y']+rect['h'])),  # right
        ((rect['x']+rect['w'], rect['y']+rect['h']), (rect['x'], rect['y']+rect['h'])),  # bottom
        ((rect['x'], rect['y']+rect['h']), (rect['x'], rect['y'])),  # left
    ]
    for C, D in rect_edges:
        if segments_intersect(A, B, C, D):
            return True
    return False

def ccw(P, Q, R):
    return (R[1]-P[1])*(Q[0]-P[0]) > (Q[1]-P[1])*(R[0]-P[0])

def segments_intersect(A, B, C, D):

    return (ccw(A, C, D) != ccw(B, C, D)) and (ccw(A, B, C) != ccw(A, B, D))


def keepout_zone_for_usb(usb_rect):

    cx, cy = center_of(usb_rect)
    # Top edge
    if abs(usb_rect['y'] - 0) < 1e-6:
        x0 = max(0, cx - KEEPOUT_WIDTH/2.0)
        return rect_from_xywh(x0, 0, min(KEEPOUT_WIDTH, BOARD_W), KEEPOUT_DEPTH)
    # Bottom
    if abs((usb_rect['y'] + usb_rect['h']) - BOARD_H) < 1e-6:
        x0 = max(0, cx - KEEPOUT_WIDTH/2.0)
        return rect_from_xywh(x0, BOARD_H - KEEPOUT_DEPTH, min(KEEPOUT_WIDTH, BOARD_W), KEEPOUT_DEPTH)
    # Left
    if abs(usb_rect['x'] - 0) < 1e-6:
        y0 = max(0, cy - KEEPOUT_WIDTH/2.0)
        return rect_from_xywh(0, y0, KEEPOUT_DEPTH, min(KEEPOUT_WIDTH, BOARD_H))
    # Right
    # else:
    x0 = BOARD_W - KEEPOUT_DEPTH
    y0 = max(0, cy - KEEPOUT_WIDTH/2.0)
    return rect_from_xywh(x0, y0, KEEPOUT_DEPTH, min(KEEPOUT_WIDTH, BOARD_H))


def place_on_edge_random(size_w, size_h, edge):
    """Return an integer (x,y) such that rect of size fits and touches the given edge.
       edge in {'top','bottom','left','right'}.
    """
    if edge == 'top':
        y = 0
        x = random.randint(0, BOARD_W - size_w)
    elif edge == 'bottom':
        y = BOARD_H - size_h
        x = random.randint(0, BOARD_W - size_w)
    elif edge == 'left':
        x = 0
        y = random.randint(0, BOARD_H - size_h)
    else:  # right
        x = BOARD_W - size_w
        y = random.randint(0, BOARD_H - size_h)
    return x, y

def random_inside_grid(size_w, size_h, margin=0):
    x = random.randint(margin, BOARD_W - size_w - margin)
    y = random.randint(margin, BOARD_H - size_h - margin)
    return x, y


def validate_all(placement):
    """Check all hard constraints. Return (is_valid, list_of_issues)."""
    issues = []

    for name, rect in placement.items():
        if not inside_board(rect):
            issues.append(f"{name} not inside board")
    # overlaps
    names = list(placement.keys())
    for i in range(len(names)):
        for j in range(i+1, len(names)):
            if rects_overlap(placement[names[i]], placement[names[j]]):
                issues.append(f"{names[i]} overlaps {names[j]}")

    def touches_edge(r):
        return (abs(r['x'] - 0) < 1e-6 or abs(r['y'] - 0) < 1e-6 or
                abs(r['x'] + r['w'] - BOARD_W) < 1e-6 or abs(r['y'] + r['h'] - BOARD_H) < 1e-6)
    for name in ['USB_CONNECTOR', 'MIKROBUS_CONNECTOR_1', 'MIKROBUS_CONNECTOR_2']:
        if name not in placement or not touches_edge(placement[name]):
            issues.append(f"{name} must touch a board edge")
    # MB1 and MB2 opposite edges and parallel orientation
    mb1 = placement.get('MIKROBUS_CONNECTOR_1')
    mb2 = placement.get('MIKROBUS_CONNECTOR_2')
    if mb1 and mb2:

        same_orientation = (abs(mb1['w'] - COMP_SIZES['MIKROBUS_CONNECTOR_1'][0]) < 1e-6) == (abs(mb2['w'] - COMP_SIZES['MIKROBUS_CONNECTOR_2'][0]) < 1e-6)
        # detect edges touched
        def which_edge(r):
            if abs(r['y'] - 0) < 1e-6: return 'top'
            if abs(r['y'] + r['h'] - BOARD_H) < 1e-6: return 'bottom'
            if abs(r['x'] - 0) < 1e-6: return 'left'
            if abs(r['x'] + r['w'] - BOARD_W) < 1e-6: return 'right'
            return None
        e1 = which_edge(mb1); e2 = which_edge(mb2)
        opposite = ((e1 == 'top' and e2 == 'bottom') or (e1 == 'bottom' and e2 == 'top') or
                    (e1 == 'left' and e2 == 'right') or (e1 == 'right' and e2 == 'left'))
        if not opposite:
            issues.append("MIKROBUS connectors must be on opposite edges")
        if not same_orientation:
            issues.append("MIKROBUS connectors must be parallel (same orientation)")
    else:
        issues.append("MIKROBUS connectors missing")

    if 'CRYSTAL' in placement and 'MICROCONTROLLER' in placement:
        d = distance(center_of(placement['CRYSTAL']), center_of(placement['MICROCONTROLLER']))
        if d > CRYSTAL_MAX_DIST + 1e-9:
            issues.append(f"CRYSTAL too far from MICROCONTROLLER ({d:.2f} > {CRYSTAL_MAX_DIST})")
    else:
        issues.append("CRYSTAL or MICROCONTROLLER missing")

    if 'USB_CONNECTOR' in placement and 'CRYSTAL' in placement and 'MICROCONTROLLER' in placement:
        keepout = keepout_zone_for_usb(placement['USB_CONNECTOR'])
        A = center_of(placement['CRYSTAL']); B = center_of(placement['MICROCONTROLLER'])
        if seg_intersects_rect(A, B, keepout):
            issues.append("Crystal-MCU line crosses USB keep-out zone")
    else:
        issues.append("USB or CRYSTAL or MICROCONTROLLER missing for keepout check")
    # global COM constraint
    centers = [center_of(r) for r in placement.values()]
    com_x = sum(c[0] for c in centers) / len(centers)
    com_y = sum(c[1] for c in centers) / len(centers)
    if distance((com_x, com_y), BOARD_CENTER) > COM_MAX_DIST + 1e-9:
        issues.append(f"Center of mass too far from board center ({distance((com_x,com_y),BOARD_CENTER):.2f} > {COM_MAX_DIST})")
    return (len(issues) == 0, issues)


def find_valid_placement(max_attempts=4000, seed=None):
    if seed is not None:
        random.seed(seed)
    start_time = time.time()
    attempts = 0

    edges = ['top', 'bottom', 'left', 'right']


    while attempts < max_attempts and (time.time() - start_time) < 1.8:
        attempts += 1
        placement = {}

        # 1) Choose MB1 edge and MB2 opposite
        mb1_edge = random.choice(edges)
        if mb1_edge == 'top': mb2_edge = 'bottom'
        elif mb1_edge == 'bottom': mb2_edge = 'top'
        elif mb1_edge == 'left': mb2_edge = 'right'
        else: mb2_edge = 'left'


        base_w, base_h = COMP_SIZES['MIKROBUS_CONNECTOR_1']
        # If mb on top/bottom, we prefer orientation with long side along x (rotate -> w=15,h=5)
        if mb1_edge in ('top','bottom'):
            mb_w, mb_h = base_h, base_w  # rotate to make length along x
        else:
            mb_w, mb_h = base_w, base_h  # keep length along y

        # Place MB1 and MB2 on respective edges at random positions
        x1, y1 = place_on_edge_random(mb_w, mb_h, mb1_edge)
        x2, y2 = place_on_edge_random(mb_w, mb_h, mb2_edge)
        # Cast to integers (grid)
        placement['MIKROBUS_CONNECTOR_1'] = rect_from_xywh(int(x1), int(y1), int(mb_w), int(mb_h))
        placement['MIKROBUS_CONNECTOR_2'] = rect_from_xywh(int(x2), int(y2), int(mb_w), int(mb_h))


        usb_w, usb_h = COMP_SIZES['USB_CONNECTOR']

        usb_ok = False
        for _ in range(25):
            usb_edge = random.choice(edges)
            ux, uy = place_on_edge_random(usb_w, usb_h, usb_edge)
            usb_rect = rect_from_xywh(int(ux), int(uy), usb_w, usb_h)
            if not rects_overlap(usb_rect, placement['MIKROBUS_CONNECTOR_1']) and not rects_overlap(usb_rect, placement['MIKROBUS_CONNECTOR_2']):
                placement['USB_CONNECTOR'] = usb_rect
                usb_ok = True
                break
        if not usb_ok:
            continue  # try another MB placement

        # 3) Place MICROCONTROLLER: sample near center first to help COM
        mc_w, mc_h = COMP_SIZES['MICROCONTROLLER']
        mc_placed = False
        # Attempt prefer center; sample points in a small radius around center first
        center_radii = [0, 2, 4, 6, 10, 15]
        for radius in center_radii:
            # sample several candidates at this radius
            samples = 20 if radius < 10 else 40
            for _ in range(samples):
                # pick offset from center uniformly in a square of side 2*radius
                if radius == 0:
                    mx = int(BOARD_CENTER[0] - mc_w/2)
                    my = int(BOARD_CENTER[1] - mc_h/2)
                else:
                    offset_x = random.randint(-radius, radius)
                    offset_y = random.randint(-radius, radius)
                    mx = int(BOARD_CENTER[0] - mc_w/2 + offset_x)
                    my = int(BOARD_CENTER[1] - mc_h/2 + offset_y)
                # clamp
                mx = max(0, min(BOARD_W - mc_w, mx))
                my = max(0, min(BOARD_H - mc_h, my))
                mc_rect = rect_from_xywh(mx, my, mc_w, mc_h)
                # avoid overlaps with MBs and USB
                if rects_overlap(mc_rect, placement['MIKROBUS_CONNECTOR_1']) or rects_overlap(mc_rect, placement['MIKROBUS_CONNECTOR_2']) or rects_overlap(mc_rect, placement['USB_CONNECTOR']):
                    continue
                placement['MICROCONTROLLER'] = mc_rect
                mc_placed = True
                break
            if mc_placed:
                break
        if not mc_placed:
            continue


        crystal_w, crystal_h = COMP_SIZES['CRYSTAL']
        mc_center = center_of(placement['MICROCONTROLLER'])
        keepout = keepout_zone_for_usb(placement['USB_CONNECTOR'])
        crystal_ok = False
        # generate candidate centers within circle radius 10 (integer)
        candidate_centers = []
        R = int(math.floor(CRYSTAL_MAX_DIST))
        for dx in range(-R, R+1):
            for dy in range(-R, R+1):
                if dx*dx + dy*dy <= CRYSTAL_MAX_DIST*CRYSTAL_MAX_DIST:
                    cx = int(round(mc_center[0] + dx))
                    cy = int(round(mc_center[1] + dy))
                    # convert center to top-left x,y for crystal (centered)
                    topx = cx - crystal_w//2
                    topy = cy - crystal_h//2
                    # ensure integer within board
                    if topx < 0 or topy < 0 or topx + crystal_w > BOARD_W or topy + crystal_h > BOARD_H:
                        continue
                    rect = rect_from_xywh(int(topx), int(topy), crystal_w, crystal_h)
                    # avoid overlap with existing
                    if rects_overlap(rect, placement['MIKROBUS_CONNECTOR_1']) or rects_overlap(rect, placement['MIKROBUS_CONNECTOR_2']) or rects_overlap(rect, placement['USB_CONNECTOR']) or rects_overlap(rect, placement['MICROCONTROLLER']):
                        continue
                    # check line intersection with keepout
                    A = center_of(rect); B = mc_center
                    if seg_intersects_rect(A, B, keepout):
                        continue
                    candidate_centers.append(rect)
        if not candidate_centers:
            continue

        candidate_centers.sort(key=lambda r: (distance(center_of(r), mc_center), distance(center_of(r), BOARD_CENTER)))
        placement['CRYSTAL'] = candidate_centers[0]
        crystal_ok = True


        valid, issues = validate_all(placement)
        if valid:

            return placement, attempts, time.time() - start_time

    return None, attempts, time.time() - start_time

placement, attempts, elapsed = find_valid_placement(max_attempts=8000, seed=42)
print("Attempts:", attempts, "Elapsed time: {:.3f}s".format(elapsed))
if placement is None:
    print("No valid placement found within limits.")
else:
    print("Found valid placement!")

    # Print components and centers
    for name, r in placement.items():
        print(f"{name}: x={r['x']}, y={r['y']}, w={r['w']}, h={r['h']}, center={center_of(r)}")

    # compute COM
    centers = [center_of(r) for r in placement.values()]
    com_x = sum(c[0] for c in centers)/len(centers)
    com_y = sum(c[1] for c in centers)/len(centers)
    print("COM:", (com_x, com_y), "distance to board center:", distance((com_x,com_y), BOARD_CENTER))

    # verify constraints again and show any messages
    valid, issues = validate_all(placement)
    print("Valid:", valid)
    if not valid:
        print("Issues:", issues)


    fig, ax = plt.subplots(figsize=(6,6))
    # board border
    ax.add_patch(plt.Rectangle((0,0), BOARD_W, BOARD_H, fill=False, linewidth=2))
    # draw keepout zone
    keepout = keepout_zone_for_usb(placement['USB_CONNECTOR'])
    ax.add_patch(plt.Rectangle((keepout['x'], keepout['y']), keepout['w'], keepout['h'], fill=True, alpha=0.18, color='red', label='USB keepout'))
    # draw components
    colors = {
        'USB_CONNECTOR': 'orange',
        'MICROCONTROLLER': 'blue',
        'MIKROBUS_CONNECTOR_1': 'green',
        'MIKROBUS_CONNECTOR_2': 'green',
        'CRYSTAL': 'purple',
    }
    for name, r in placement.items():
        ax.add_patch(plt.Rectangle((r['x'], r['y']), r['w'], r['h'], alpha=0.7, color=colors.get(name, 'gray')))
        cx, cy = center_of(r)
        ax.text(cx, cy, name, ha='center', va='center', fontsize=8, color='white' if name!='USB_CONNECTOR' else 'black')


    mc_center = center_of(placement['MICROCONTROLLER'])
    circle = plt.Circle(mc_center, CRYSTAL_MAX_DIST, fill=False, linestyle='--', linewidth=1)
    ax.add_patch(circle)
    # draw COM
    ax.plot(com_x, com_y, marker='x', color='black')
    ax.text(com_x+0.8, com_y+0.8, 'COM', fontsize=8)
    ax.set_xlim(-1, BOARD_W+1)
    ax.set_ylim(-1, BOARD_H+1)
    ax.set_aspect('equal')
    ax.invert_yaxis()  # follow top-left origin with y downwards
    ax.set_title(f"Found placement (COM dist {distance((com_x,com_y),BOARD_CENTER):.2f})")
    plt.show()
```
